# Data Visualization in Python with Matplotlib! 

### Load necessary libaries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your graph (dpi specifies pixels per inch. when saving probably should be 300 if possible)
plt.figure(figsize =(5,3), dpi=100)


#Line 1

# Keyword argument notation
# plt.plot(x,y, label = '2x', color= 'green', linewidth = 3, marker = '.', markersize =20, markeredgecolor ='red',  linestyle = '-.' )

# Sorthand notataion
plt.plot(x,y, 'go--',label = '2x')


## Line 2

#select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label ='X^2')

# Plot remainder of the graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')


# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph !', fontdict={'fontname' : 'Comic Sans MS','fontsize': 20})

# X and Y labels
plt.xlabel('X - Axis')
plt.ylabel('y - Axis')


# X , Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
# plt.yticks([0,2,4,6,7,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resulation)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]


bars = plt.bar(labels, values)

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4))

plt.show()
```
